```python
import glob
import os
import numpy as np
results_path = '/mnt/storage/bender/projects/cl/sigmorphon-uzh/paper2018/results/'

for dataset in ['newcelex', 'sgm2017low', 'sgm2017medium']:
    for mode in ['rl', 'mrt', 'rlnb', 'mrt1', 'mrt01']:
        max_epochs = []
        for log_fn in glob.glob(os.path.join(results_path, dataset, 'x-*-m%s-x/*/*/f.log' % mode)):
            #print log_fn
            max_epoch = -1
            max_acc = 0
            with open(log_fn) as f:
                for line in f:
                    if line.startswith('epoch'):
                        max_epoch = -1
                        max_acc = 0
                        continue
                    line = line.strip().split('\t')
                    #print line
                    if len(line) == 4:
                        epoch, _, _, dev_acc = line
                        epoch = int(epoch)
                        dev_acc = float(dev_acc)
                        if dev_acc > max_acc:
                            max_epoch = epoch
                            max_acc = dev_acc
            #print max_epoch, max_acc
            max_epochs.append(max_epoch)
        if max_epochs:
            print 'DATASET, MODE, MEAN EPOCH (STD): %s\t%s\t%.3f\t%.3f' % (dataset,
                   mode, np.mean(max_epochs), np.std(max_epochs))
```

```python
def is_outlier(points, thresh=3.5):
    """
    Returns a boolean array with True if points are outliers and False 
    otherwise.

    Parameters:
    -----------
        points : An numobservations by numdimensions array of observations
        thresh : The modified z-score to use as a threshold. Observations with
            a modified z-score (based on the median absolute deviation) greater
            than this value will be classified as outliers.

    Returns:
    --------
        mask : A numobservations-length boolean array.

    References:
    ----------
        Boris Iglewicz and David Hoaglin (1993), "Volume 16: How to Detect and
        Handle Outliers", The ASQC Basic References in Quality Control:
        Statistical Techniques, Edward F. Mykytka, Ph.D., Editor. 
    """
    if len(points.shape) == 1:
        points = points[:,None]
    median = np.median(points, axis=0)
    diff = np.sum((points - median)**2, axis=-1)
    diff = np.sqrt(diff)
    med_abs_deviation = np.median(diff)

    modified_z_score = 0.6745 * diff / med_abs_deviation

    return modified_z_score > thresh
```

```python
class Config(object):
    def __init__(self, name, num_folds=5, num_seeds=5):
        self.name = name
        self.num_folds = num_folds
        self.num_seeds = num_seeds
        self.accs = dict()  # condition (dev/test), decoding (greedy, '', beam*)
    def __getitem__(self, key):
        key = tuple(key)
        if key in self.accs:        
            item = self.accs[key]
        else:
            item = np.zeros((self.num_folds, self.num_seeds))
            self.accs[key] = item
        return item
```
